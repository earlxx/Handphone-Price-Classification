# Modeling

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns='price_range')
y = df.price_range

X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,stratify=y,random_state=0)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# mengimport library dan algoritma

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

## KNN

```python
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
knn.score(X_test,y_test),knn.score(X_train,y_train)
```

Hasil KNN dengan neighbors = 5 terlihat score test sebesar 91%, dan score train sebesar 94%.
ini menunjukan bahwa model kita tidak overfit

## Random Forest

```python
# Pipeline
numeric_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler',MinMaxScaler())
])

categori_pipe = Pipeline([
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('onehot',OneHotEncoder())
])

# Preprocessor
from sklearn.compose import ColumnTransformer
preproses = ColumnTransformer([
    ('numeric',numeric_pipe,['battery_power','clock_speed','fc','int_memory',
                              'm_dep','mobile_wt','n_cores','pc','px_height','px_width','ram','sc_h','sc_w','talk_time']),
    ('kategori',categori_pipe,['blue','dual_sim','four_g','three_g','touch_screen','wifi'])
])

# Algo Pipeline KNN 
pipeline = Pipeline([
    ('preproses',preproses),
    ('algo',RandomForestClassifier())
])

# PARAMS KNN
params_rf = {
    'algo__n_estimators':[100,150,200],
    'algo__max_depth':[20,50,80],
    'algo__max_features':[0.3,0.6,0.8],
    'algo__min_samples_leaf':[1,5,10]
}


model = RandomizedSearchCV(pipeline,params_rf,cv=3,n_iter=50,n_jobs=-1,verbose=1,random_state=42)
model.fit(X_train,y_train)
print(model.best_params_)
print(model.score(X_train,y_train),model.score(X_test,y_test))
```

tuning menggunakan random foret, skor train sebesar 100% , skor test sebesar 91% dengan parameter terbaik adalah n_estimators = 100, min_samples_leaf = 1, max_features = 0.8, max-depth = 50

```python
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
np.argmax(rf.feature_importances_)
tree = pd.Series(rf.feature_importances_,X.columns)
plt.figure(figsize=(10,6))
# plt.axis([0,1,None,None])
plt.barh(X.columns,tree)
```

melihat Feature Importance, bahwa RAM tertinggi signifikan disusul dengan battery power dan ukuran.

```python
# Pipeline
numeric_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler',MinMaxScaler())
])

categori_pipe = Pipeline([
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('onehot',OneHotEncoder())
])

# Preprocessor
from sklearn.compose import ColumnTransformer
preproses = ColumnTransformer([
    ('numeric',numeric_pipe,['battery_power','clock_speed','fc','int_memory',
                              'm_dep','mobile_wt','n_cores','pc','px_height','px_width','ram','sc_h','sc_w','talk_time']),
    ('kategori',categori_pipe,['blue','dual_sim','four_g','three_g','touch_screen','wifi'])
])

# Algo Pipeline KNN 
pipeline = Pipeline([
    ('preproses',preproses),
    ('algo',RandomForestClassifier())
])

# PARAMS rf
params_rf = {
    'algo__n_estimators':[100,150,200],
    'algo__max_depth':[20,50,80],
    'algo__max_features':[0.3,0.6,0.8],
    'algo__min_samples_leaf':[1,5,10]
}

model = RandomizedSearchCV(pipeline,params_rf,cv=3,n_iter=50,n_jobs=-1,verbose=1,random_state=42)
model.fit(X_train,y_train)
print(model.best_params_)
# print(model.score(X_train,y_train),model.best_score_,model.score(X_test,y_test))
print ("FINAL SCORE : ", format(model.score(X_test,y_test)*100,'.1f'),'%')
```
